#LOWERCASING OF SENTENCES

```python
sentence = "My name is Saurabh Singh"
```

```python
lowercase_sentence = sentence.lower()
```

```python
print(lowercase_sentence)
```

```python
sentence_list = ['The sun shines.',
                 'Birds fly in the sky.',
                 'She likes to read.',
                 'I have a pen.',
                 'The baby is sleeping',]
```

```python
lowe_sentence_list = [x.lower() for x in sentence_list] 
```

```python
print(lowe_sentence_list)
```
