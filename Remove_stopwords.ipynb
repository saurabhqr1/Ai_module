#REMOVING STOP WORDS like "AND" "OF" "A" "to"


```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
```

```python
print(en_stopwords)
```

```python
sentence = "it was too far to go to the shop and he did not want her to walk"
```

```python
sentence_no_stopword = ' '.join([word for word in sentence.split() if word not in (en_stopwords)])
print(sentence_no_stopword)
```
