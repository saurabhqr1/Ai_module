## Problem
Give  Daily Sales of a shop for 7 days.
[1200, 1800, 1500, 7000, 8000, 1700, 5500]
1. find total sale (week)
2. Average sale
3. highest sale and lowest sales
4. day no. when the highest sales happen

```python
import numpy as np
ar1=np.array([1200, 1800, 1500, 7000, 8000, 1700, 5500])
print(ar1)
total_sale = np.sum(ar1)
print("Total Sale(week) :" , total_sale)
high_sale = np.max(ar1)
print("Max sale :", high_sale)
low_sale = np.min(ar1)
print("Min Sale :", low_sale)
avrage_sale = np.mean(ar1)
print("Avg Sale :",avrage_sale)
day_high = np.argmax(ar1) +1
print("Highest Sale Day :",day_high)
```
